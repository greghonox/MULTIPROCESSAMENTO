```python
from concurrent import futures
from requests import get, post
from queue import Queue
```

```python
url = 'https://webhook.site/a675ad23-057f-4545-bcd9-c4366dd7e7d2'
```

```python
WORKS = 100
SOLIC = 500
solicity = Queue()
for _ in range(SOLIC): 
    solicity.put(f"AMO JESUS {_}")
```

```python
def get_d():
    return get(url)

def post_d(job, r):
    print(f'WORK EXECUTANDO {r} {job}')
    return post(url, data={'name': 'AMO JESUS', 'job': job, 'work': r})
```

```python
def request(r):
    while not solicity.empty():
        job = solicity.get()
        post_d(job, r)
```

```python
with futures.ThreadPoolExecutor(max_workers=WORKS) as pool:
    threads = []
    for work in range(WORKS):
        threads.append(pool.submit(request,
                                   work))
    
    futures.wait(threads, return_when='ALL_COMPLETED')
print("[[main]] &gt;&gt; Processo finalizado")
```
